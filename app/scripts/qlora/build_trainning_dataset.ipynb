```python
# instruction
"""
Act as a coding learning test assistant. 
Analyze the coding problem step by step for a learner. 
Explain the problem understanding, input/output, core concept, solution strategy, implementation plan, and verification points.
"""
df = df.drop(['instruction', 'metadata'], axis=1)
df.head()
```

```python
prompt = """

# Coding Test Problem Analyzer

You are an expert algorithm instructor and competitive programming mentor.

Your task is to analyze 4 coding interview / algorithm problems written in English and convert each of them into a structured JSON format, then return them as a JSON list in the same order as the input.

Requirements:

1. Translate the original problem into natural Korean (problem_korean field only).
2. ALL OTHER FIELDS MUST BE IN ENGLISH. Do not use Korean anywhere except in the problem_korean field.
3. Rewrite the problem in simpler English so that a beginner can understand it easily.
4. Identify the most appropriate algorithm(s) or data structure(s) needed to solve the problem.
5. Explain why those algorithms were selected.
6. Create a high-level implementation plan without writing code.
7. Do NOT provide code.
8. Do NOT provide the final answer.
9. Return a JSON array only, containing exactly 4 objects, one per input problem, in the same order as given.
10. Every field must contain meaningful content.
11. algorithm_selection must contain algorithm or data structure names only.
12. implementation_plan should describe the overall solution flow, not code details.

Output Schema (array of 4 objects):

[
  {{
    "problem_korean": "",
    "problem_understanding": "",
    "algorithm_selection": [],
    "selection_reason": "",
    "implementation_plan": ""
  }}
]

Field Definitions:

- problem_korean

  - Natural Korean translation of the entire problem.
  - Preserve all requirements and constraints.
  - This is the ONLY field that should contain Korean text.

- problem_understanding

  - Simplified explanation of the problem in English.
  - Rewrite the problem in a clearer and easier-to-understand form.
  - Focus on what must be achieved and what the input/output represent.

- algorithm_selection

  - List of the most appropriate algorithms or data structures.
  - Written in English.
  - Examples:
    - Array
    - Hash Map
    - Two Pointers
    - Sliding Window
    - BFS
    - DFS
    - Dynamic Programming
    - Greedy
    - Binary Search
    - Graph
    - Heap
    - Sorting

- selection_reason

  - Explain why the selected algorithm(s) fit this problem.
  - Mention important observations, constraints, and characteristics of the problem.
  - Explain why simpler or naive approaches may be inefficient if applicable.

- implementation_plan

  - High-level solution plan written in English.
  - Describe the sequence of steps an experienced programmer would follow.
  - Focus on logic and strategy.
  - Do NOT write code.
  - Do NOT mention programming language syntax.

Input Problems:

{PROBLEMS}

---

Return a JSON array of 4 objects only.
No markdown.
No explanation.
No additional text.
"""
```

```python
from dotenv import load_dotenv
from openai import OpenAI
import pandas as pd
import json
import os

load_dotenv()

redo_list = [
    108 ,320 ,792  ,824 ,880 
]

client = OpenAI(
    api_key=os.getenv("OPENAI_API_KEY")
)

questions = df["input"].tolist()

OUTPUT_FILE = "algorithm_dataset.csv"
SAVE_EVERY_BATCH = 10

# 새로 시작할 경우 기존 파일 삭제
if os.path.exists(OUTPUT_FILE):
    os.remove(OUTPUT_FILE)

results = []

for start in redo_list:

    batch = questions[start:start + 4]

    if len(batch) < 4:
        break

    formatted_prompt = prompt.format(
        PROBLEMS="\n\n".join(
            f"### Problem {i + 1}\n{problem}"
            for i, problem in enumerate(batch)
        )
    )

    try:

        response = client.responses.create(
            model="gpt-5",
            input=formatted_prompt
        )

        parsed = json.loads(response.output_text)

        for original_question, analyzed in zip(batch, parsed):

            results.append({
                "question": original_question,
                "problem_korean": analyzed["problem_korean"],
                "problem_understanding": analyzed["problem_understanding"],
                "algorithm_selection": json.dumps(
                    analyzed["algorithm_selection"],
                    ensure_ascii=False
                ),
                "selection_reason": analyzed["selection_reason"],
                "implementation_plan": analyzed["implementation_plan"]
            })

        current_batch = start // 4 + 1

        # print(
        #     f"[OK] batch={current_batch} "
        #     f"processed={min(start + 4, len(questions))}/{len(questions)}"
        # )

        # batch마다 append 저장
        if current_batch % SAVE_EVERY_BATCH == 0:

            temp_df = pd.DataFrame(results)

            temp_df.to_csv(
                OUTPUT_FILE,
                mode="a",
                header=not os.path.exists(OUTPUT_FILE),
                index=False,
                encoding="utf-8-sig"
            )

            print(
                f"[SAVE] batch={current_batch} "
                f"rows={len(temp_df)}"
            )

            results.clear()

    except Exception as e:

        print(
            f"[ERROR] batch={start // 4 + 1} "
            f"start={start} "
            f"error={e}"
        )

# 마지막 남은 데이터 저장
if results:

    temp_df = pd.DataFrame(results)

    temp_df.to_csv(
        OUTPUT_FILE,
        mode="a",
        header=not os.path.exists(OUTPUT_FILE),
        index=False,
        encoding="utf-8-sig"
    )

    print(
        f"[FINAL SAVE] rows={len(temp_df)}"
    )

print("Done")
```
